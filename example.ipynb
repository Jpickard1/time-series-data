# 2018, Myogenic Reprogramming

```python
import pickle
import numpy

file_path = "data/2018_data.pkl"
with open(file_path, "rb") as f:
    data = pickle.load(f)

# Inspect the loaded data
print(type(data))  # Check the data type
print(data.keys() if isinstance(data, dict) else data[:5])  # Preview content if it's a dict or list
print(data.keys() if isinstance(data, dict) else data[:5])  # Preview content if it's a dict or list
print(f"{data['tpm'].shape=}")
data['replicates']['r1']
```
